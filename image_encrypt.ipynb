```python
from PIL import Image
import numpy as np

img=Image.open('img.jpeg')
img_array=np.array(img)
img_array
```

```python
import matplotlib.pyplot as plt
plt.imshow(img_array)
plt.axis('off')
plt.show()
```

```python
from blowfish_1 import Blowfish
```

```python
blowfish_img=Blowfish("ajdshajksdhf")
blowfish_img.subkeyGeneration()
```

```python
encrypted_array = np.zeros_like(img_array, dtype=np.uint64)
for i in range(img_array.shape[0]):
    for j in range(img_array.shape[1]):
        for k in range(img_array.shape[2]):
            encrypted_array[i][j][k]=(blowfish_img.blowFish_encrypt(int(img_array[i][j][k])))
```

```python
encrypted_array
```

```python

arr=encrypted_array%255
plt.imshow(arr)
plt.axis('off')
plt.show()
```
